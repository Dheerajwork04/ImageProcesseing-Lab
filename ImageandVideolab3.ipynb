```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Step 1: Create the 2D unit step signal g1(x, y)
size = 150
g1 = np.zeros((size, size))

# Modify to have half black and half white
g1[:, size//2:] = 1

# Step 2: Visualize the 2D unit step signal in 2D
plt.figure(figsize=(4, 4))  # Reduced figure size
plt.imshow(g1, cmap='gray')
plt.title('2D Unit Step Signal')
plt.colorbar()
plt.show()

# Step 3: Visualize the 2D unit step signal in 3D
x = np.linspace(0, size-1, size)
y = np.linspace(0, size-1, size)
X, Y = np.meshgrid(x, y)
fig = plt.figure(figsize=(6, 6))  # Reduced figure size
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, g1, cmap='viridis')
ax.set_title('3D View of 2D Unit Step Signal')
plt.show()

# Step 4: Compute the 2D DFT of g1(x, y)
G1 = np.fft.fftshift(np.fft.fft2(g1))

# Step 5: Compute the magnitude spectrum
magnitude_spectrum = np.abs(G1)

# Step 6: Visualize the magnitude spectrum in 2D
plt.figure(figsize=(4, 4))  # Reduced figure size
plt.imshow(np.log(1 + magnitude_spectrum), cmap='gray')
plt.title('2D Magnitude Spectrum of G1(u, v)')
plt.colorbar()
plt.show()

# Step 7: Visualize the magnitude spectrum in 3D
fig = plt.figure(figsize=(6, 6))  # Reduced figure size
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, np.log(1 + magnitude_spectrum), cmap='viridis')
ax.set_title('3D View of Magnitude Spectrum of G1(u, v)')
plt.show()
```

```python
#2.
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Step 1: Create the 2D impulse signal g2(x, y)
size = 150
impulse_size = 10
g2 = np.zeros((size, size))

# Determine the center of the 2D array
center_x, center_y = size // 2, size // 2

# Create a 10x10 impulse at the center
g2[center_x - impulse_size//2 : center_x + impulse_size//2, 
   center_y - impulse_size//2 : center_y + impulse_size//2] = 1

# Step 2: Visualize the 2D impulse signal in 2D
plt.figure(figsize=(4, 4))  # Reduced figure size
plt.imshow(g2, cmap='gray')
plt.title('2D Impulse Signal (10x10 Impulse)')
plt.colorbar()
plt.show()

# Step 3: Visualize the 2D impulse signal in 3D
x = np.linspace(0, size-1, size)
y = np.linspace(0, size-1, size)
X, Y = np.meshgrid(x, y)
fig = plt.figure(figsize=(6, 6))  # Reduced figure size
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, g2, cmap='viridis')
ax.set_title('3D View of 2D Impulse Signal')
plt.show()

# Step 4: Compute the 2D DFT of g2(x, y)
G2 = np.fft.fftshift(np.fft.fft2(g2))

# Step 5: Compute the magnitude spectrum
magnitude_spectrum = np.abs(G2)

# Step 6: Visualize the magnitude spectrum in 2D
plt.figure(figsize=(4, 4))  # Reduced figure size
plt.imshow(np.log(1 + magnitude_spectrum), cmap='gray')
plt.title('2D Magnitude Spectrum of G2(u, v)')
plt.colorbar()
plt.show()

# Step 7: Visualize the magnitude spectrum in 3D
fig = plt.figure(figsize=(6, 6))  # Reduced figure size
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, np.log(1 + magnitude_spectrum), cmap='viridis')
ax.set_title('3D View of Magnitude Spectrum of G2(u, v)')
plt.show()
```

```python
#4
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy.fftpack import dct, idct

# Step 1: Load the grayscale image
image_path = r'C:\Users\DELL\Downloads\sunflower_gray.jpg'  # Update this path to your image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Ensure the image is correctly loaded
if image is None:
    raise ValueError("Image not found or path is incorrect. Please check the image path.")

# Convert to float32 for DCT computation
image = image.astype(np.float32)

# Step 2: Compute the 2D DCT of the image
dct_image = dct(dct(image.T, norm='ortho').T, norm='ortho')

# Step 3: Reconstruct the image using the inverse DCT
reconstructed_image = idct(idct(dct_image.T, norm='ortho').T, norm='ortho')

# Clip values to the valid range [0, 255] and convert back to uint8
reconstructed_image = np.clip(reconstructed_image, 0, 255)
reconstructed_image = reconstructed_image.astype(np.uint8)

# Step 4: Display the original, DCT, and reconstructed images
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.colorbar()

plt.subplot(1, 3, 2)
plt.imshow(np.log(1 + np.abs(dct_image)), cmap='gray')
plt.title('DCT of Image')
plt.colorbar()

plt.subplot(1, 3, 3)
plt.imshow(reconstructed_image, cmap='gray')
plt.title('Reconstructed Image')
plt.colorbar()

plt.show()
```
